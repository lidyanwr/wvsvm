```python
from bs4 import BeautifulSoup
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords 
import math
lemmatizer=WordNetLemmatizer()

dictkeyword = [line.rstrip('\n') for line in open('file/keyword/unique_keyword.txt')]


def remove_whitespace(text, dictkeyword=None): 
    return  " ".join(text.split()) 

def keyword(raw_text):
    all_words = raw_text.split()
    mykeyword = [word for word in dictkeyword if word in all_words]
    return (mykeyword)
    
def varkeyword (raw_text):
    keywordvar = len(raw_text)
    return (keywordvar)

def freq_keyword (raw_text):
    all_words = raw_text.split()
    freqkey = [word for word in all_words if word in dictkeyword]
    freqKeyword = len(freqkey)
    return (freqKeyword)

def term_count(raw_text):
    allwords = raw_text.split()
    filtered_text = [word for word in allwords]
    counterm = len(filtered_text)
    return(counterm)

allkeyword = []
frequency = {}
myword = []
freqword = []

def term_doc_count(key):
    for line in key:
        for x in line:
            allkeyword.append(x)
    #for x in myterm:
    jumlah = len(allkeyword)
    return (jumlah)

def freq_doc(mylist):
    for word in mylist:
        count = frequency.get(word,0)
        frequency[word] = count + 1
    frequency_list = frequency.keys()
     
    for words in frequency_list:
        myword.append(words)
        freqword.append(frequency[words])
    return 0

def findInterval(key):
    listAngka = []
    numbkey = len(key)
    if numbkey != 0:
        for x in key:
            indexTerm = freqKeyword[freqKeyword['Term'] == x].index
            numberKu = freqKeyword["TotalFrequency"].loc[indexTerm].values
            listAngka.append(numberKu)
    
        kelasAtas = max(listAngka)
        kelasBawah = min(listAngka)
        A = int(''.join(str(i) for i in kelasAtas))
        B = int(''.join(str(i) for i in kelasBawah))
        myrange = A - B
        hasilog =  math.log10(numbkey)
        kelas = round(1 + (3.3 * math.log10(numbkey)))
        interval = round(myrange / kelas)
    
    else:
        interval = 0
    #print("N: ",numbkey)
    #print("hasil log: ", hasilog)
    #print("Jumlah kelas : ", kelas)
    #print("Kelas Atas: ", A)
    #print("Kelas Bawah: ", B)
    #print("Range: ", myrange)
    #print("Interval ", interval)
    #print(  )
    return (interval)
 
```
